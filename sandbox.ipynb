```python
# class UserRegistration function is to register a pair of users
class UserRegistration:
    #the __init__ method is called when the class is instantiated. Its role in the code is to initialize the class.
    def __init__(self):
        self.pairs = []
    # the registrer_pair method is called when the class is instantiated. Its role in the code is to register a pair of users.
    def register_pair(self, user1, user2):
        self.pairs.append((user1, user2))
        print("{}, {} en vous inscrivant vous avez créé un...".format(user1, user2))

```

```python
#Below the code acts as a web server that allows users to register pairs of users.
from flask import Flask, render_template, request

app = Flask(__name__)
registration = UserRegistration()

#the code below is a route that allows users to register pairs of users.
@app.route('/', methods=['GET', 'POST'])
#The def index() function is called when the class is instantiated. Its role in the code is to render the index.html template.
def index():
    error_message = ""
    user1_text = ""
    user2_text = ""
    success_message = ""
    if request.method == 'POST':
        user1 = request.form['user1']
        user2 = request.form['user2']
        
        if user1 == "" or user2 == "":
            error_message = "Ah non, pas comme ça Zinedine. Il faut être deux pour pouvoir utiliser Dataflow"
        else:
            registration.register_pair(user1, user2)
            user1_text = f"Oh non {user1}, c'est terrible !"
            user2_text = f"{user2}, ne paniquez pas mais je crois qu'il y a un bug...NON ! NOOOOOOONNNNNNNNN"
            success_message = "Bienvenue {}, {} ravi de vous savoir parmi nous.".format(user1, user2)
    return render_template('index.html', error_message=error_message, user1_text=user1_text, user2_text=user2_text, success_message=success_message)
   

#the "if __name__ == '__main__':" statement is used to check whether the current script is executed on its own or imported somewhere else by some other script.
if __name__ == '__main__':
    app.run(port=5000)  # Use any available port, such as 5000
```
